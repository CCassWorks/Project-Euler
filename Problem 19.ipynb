```python
import numpy as np

daylist = ['mon', 'tues', 'wed', 'thurs', 'fri', 'sat', 'sun']
monthlist = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
monthdays = {
    'jan':31,
    'feb':28,
    'mar':31,
    'apr':30,
    'may':31,
    'jun':30,
    'jul':31,
    'aug':31,
    'sep':30,
    'oct':31,
    'nov':30,
    'dec':31
}
```
