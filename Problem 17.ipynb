<p>If the numbers $1$ to $5$ are written out in words: one, two, three, four, five, then there are $3 + 3 + 5 + 4 + 4 = 19$ letters used in total.</p>
<p>If all the numbers from $1$ to $1000$ (one thousand) inclusive were written out in words, how many letters would be used? </p>
<br><p class="note"><b>NOTE:</b> Do not count spaces or hyphens. For example, $342$ (three hundred and forty-two) contains $23$ letters and $115$ (one hundred and fifteen) contains $20$ letters. The use of "and" when writing out numbers is in compliance with British usage.</p>

```python
def number_lister(top):
    x = []
    for i in range(1,top+1):
        match len(str(i)):
            case 1:
                x.append(single_digits(str(i)))
            case 2:
                x.append(double_digits(str(i)))
            case 3:
                x.append(hundreds(str(i)))
            case 4:
                x.append("onethousand")

    return x

def single_digits(x):
    match x:
        case '1':
            return "one"
        case '2':
            return "two"
        case '3':
            return "three"
        case '4':
            return "four"
        case '5':
            return "five"
        case '6':
            return "six"
        case '7':
            return "seven"
        case '8':
            return "eight"
        case '9':
            return "nine"
        case '0':
            return ""
        
def double_digits(x):
    match str(x)[0]:
        case '1':
            return tens(x)
        case '2':
            return "twenty"+single_digits(x[1])
        case '3':
            return "thirty"+single_digits(x[1])
        case '4':
            return "forty"+single_digits(x[1])
        case '5':
            return "fifty"+single_digits(x[1])
        case '6':
            return "sixty"+single_digits(x[1])
        case '7':
            return "seventy"+single_digits(x[1])
        case '8':
            return "eighty"+single_digits(x[1])
        case '9':
            return "ninety"+single_digits(x[1])
        case '0':
            return single_digits(x[1])

def tens(x):
    match x:
        case '10':
            return "ten"
        case '11':
            return "eleven"
        case '12':
            return "twelve"
        case '13':
            return "thirteen"
        case '14':
            return "fourteen"
        case '15':
            return "fifteen"
        case '16':
            return "sixteen"
        case '17':
            return "seventeen"
        case '18':
            return "eighteen"
        case '19':
            return "nineteen"
        
def hundreds(x):
    if str(x)[1:] != "00":
        return single_digits(str(x)[0])+"hundredand"+double_digits(str(x)[1:])
    else:
        return single_digits(str(x)[0])+"hundred"
    

nums = number_lister(1000)
```

```python
sum=0
for i in range(len(nums)):
    print(nums[i])
    sum=sum+len(nums[i])

print(sum)
```
